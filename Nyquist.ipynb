```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft

plt.style.use("https://github.com/mlefkir/beauxgraphs/raw/main/beautifulgraphs_colblind.mplstyle")
```

```python
f = np.linspace(-50,50,1000)
alpha = 10
psd = 1/(alpha**2+(2*np.pi*f)**2)
fig, ax  = plt.subplots(2,1,figsize=(8,5))
ax[0].plot(f,psd)
ax[0].set_xlim(-50,50)
ax[0].set_ylim(0)
psd_30 = 1/(alpha**2+(2*np.pi*(f-30))**2)
psd_m30 = 1/(alpha**2+(2*np.pi*(f+30))**2)
ax[0].plot(f,psd_m30,c="C1")
ax[0].plot(f,psd_30,c="C1")
ax[0].axvline(15,c="k",ls="--")
ax[0].axvline(-15,c="k",ls="--")
ax[0].axvline(45,c="k",ls="--")
ax[0].axvline(-45,c="k",ls="--")

psd_15 = 1/(alpha**2+(2*np.pi*(f-10))**2)
psd_m15 = 1/(alpha**2+(2*np.pi*(f+10))**2)
ax[1].axvline(5,c="k",ls="--")
ax[1].axvline(-5,c="k",ls="--")
ax[1].axvline(-15,c="k",ls="--")
ax[1].axvline(15,c="k",ls="--",label=r"$f_{\rm max}$")
ax[1].plot(f,psd)
ax[1].plot(f,psd_m15,c="C1")
ax[1].plot(f,psd_15,c="C1")
ax[1].set_xlim(-20,20)
ax[1].set_ylim(0)
ax[1].set_xlabel("Frequency")
ax[1].set_ylabel("Power Spectral Density")
ax[1].legend()
```
